```python
model=keras.Sequential()
```

```python
model.add(layers.Input(X.shape[1]))
```

```python
model.add(layers.Dense(4,activation='relu'))
```

```python
model.add(layers.Dense(512,activation='sigmoid'))
```

```python
model.compile(loss="mse")
```

```python
model.fit(X,y, epochs = 800 )
```
